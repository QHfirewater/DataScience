```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'PingFang HK'
```

## 基本饼图绘制

```python
fig, ax = plt.subplots(figsize=(10, 5), subplot_kw={'aspect': 'equal'})
recipe = ['375 g flour',
          '75 g sugur',
          '250 g butter',
          '300 g berriers']

data = [float(item.split()[0]) for item in recipe]
ingredients = [item.split()[-1] for item in recipe]

def func(pct, allvals):
    absolute = int(pct/100*np.sum(allvals))
    return '{:.1f}\n({:d}g)'.format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data), textprops={'color': 'w'})

ax.legend(wedges, ingredients,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Matplotlib bakery: A pie")

plt.show()
```

## 绘制拆解饼图

### 准备数据

```python
labels = ['其他', '水溶性离子', 'EC', 'OM']
data = [35.9, 42.2, 3.8, 18.1] # 求和为100
component = ['$Ca^{2+}$', '$Cl-$', '$NH_4^+$', '$K^+$', '$Mg^{2+}$', '$NO_3^-$', '$Na^+$', '$SO_4^{2-}$']
data_ = [0.7, 5, 26.5, 1.4, .2, 47, 3.5, 15.7]
```

### 开始绘图

```python
from matplotlib.patches import ConnectionPatch
# 创建画布和子图区域分布
fig = plt.figure(figsize=(12, 6), dpi=120)
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
fig.subplots_adjust(wspace=0)

# 旋转第二个水溶性离子成分到垂直
angle = -180 * np.sum(data[1]) / np.sum(data) - 360 * np.sum(data[0]) / np.sum(data)
# 绘制主要成分
colors = ['#01FFFF', '#FF0000', '#01FF00', '#1600FF']
ax1.pie(data, autopct=None, labels=['%s\n%.1f%%'%z for z in zip(labels, data)], 
        startangle=angle, colors=colors)

radius = .5
patches, texts = ax2.pie(data_, labels=['%s\n%.1f%%'%z for z in zip(component, data_)], 
                         radius=radius, textprops={'size': 'smaller'})

x, y = texts[0].get_position()
texts[0].set_position((x+0.1, y-0.1))
x, y = texts[4].get_position()
texts[4].set_position((x-0.12, y-0.1))

# 使用 ConnectionPatch 连接两个子图
# get the wedge data
theta1, theta2 = ax1.patches[1].theta1, ax1.patches[1].theta2
center, r = ax1.patches[1].center, ax1.patches[1].r

# 绘制顶部连接线
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(0, radius), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color('#4572C4')
con.set_linewidth(1)
ax2.add_artist(con)

# 绘制底部连接线
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(0, -radius), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color('#4572C4')
ax2.add_artist(con)
con.set_linewidth(1)

plt.show()
```
