```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 设置中文支持字体
plt.rcParams['font.family'] = 'PingFang HK'
```

## 准备数据集

```python
labels = ['$PM_{2.5}$', '$PM_{10}$', '$SO_2$', '$NO_2$', '$CO$', '$O_3-8H$']
data_2019 = [37, 62, 10, 26, 1.0, 171]
data_2020 = [39, 66, 10, 27, 1.0, 166]
```

## 准备条形图标注函数

```python
def autolabel(rects, ax, fmt):
    '''为每一个bar上添加标签'''
    for rect in rects:
        height = rect.get_height()
        ax.annotate(fmt.format(height), xy=(rect.get_x()+rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom')
```

## 绘制图片

```python
fig = plt.figure(figsize=(8, 6), dpi=120)
ax = fig.add_subplot(111)
ax_  = ax.twinx()
x = np.arange(len(labels))  # 设定标签位置
width = 0.35 # 柱状的宽度
bar_ticks1 = x-width/2
bar_ticks2 = x+width/2
rects1 = ax.bar(list(bar_ticks1[:4])+[bar_ticks1[5]], data_2019[:4]+[data_2019[5]], width, 
                label='2019.04', color='#FFC68C')
rects2 = ax.bar(list(bar_ticks2[:4])+[bar_ticks2[5]], data_2020[:4]+[data_2020[5]], width, 
                label='2020.04', color='#0080FF')
rects1_ = ax_.bar(bar_ticks1[4], data_2019[4], width, color='#FFC68C')
rects2_ = ax_.bar(bar_ticks2[4], data_2020[4], width, color='#0080FF')
autolabel(rects1, ax, '{:.0f}')
autolabel(rects2, ax, '{:.0f}')
autolabel(rects1_, ax_, '{:.1f}')
autolabel(rects2_, ax_, '{:.1f}')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('浓度($μg/m^3$)')
ax_.set_ylabel('CO 95百分位数($mg/m^3$)')
ax.legend()
plt.title('常规六参数同比统计图')
plt.show()
```
