```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.rcParams["font.family"] = 'PingFang HK'
```

```python
file = r'../data/city_normal.xls'
data = pd.read_excel(file, index_col=0)
data.head()
```

```python
fig = plt.figure(figsize=(16, 8), dpi=120)
ax = fig.add_subplot(111)

labels = ['市%d'%i for i in range(1, 9)]
x = data.iloc[:, 1].to_list()
l = [50]*len(x)
ax.stackplot(x, l, l, l, colors=
             [np.array([0, 176, 80])/255, np.array([254, 255, 0])/255, np.array([255, 192, 3])/255], alpha=.7)
for i in range(7):
    ax.plot(x, data.iloc[:, 5*i+2].to_numpy(), label=labels[i])
    
ax.legend(loc=9, ncol=8, fontsize=13, bbox_to_anchor=(.5, 1.07), edgecolor='#FFFFFF')
ax.set_xlim(pd.datetime(2020, 1, 1), pd.datetime(2020, 1, 31))
ax.set_ylim(0, 150)
plt.show()
```

## 设置背景色实现

```python
fig = plt.figure(figsize=(16, 8), dpi=120)
ax = fig.add_subplot(111)

for i in range(7):
    ax.plot(x, data.iloc[:, 5*i+2].to_numpy(), label=labels[i])
ax.axhspan(0, 50, facecolor=np.array([0, 176, 80])/255, alpha=.7)
ax.axhspan(50, 100, facecolor=np.array([254, 255, 0])/255, alpha=.7)
ax.axhspan(100, 150, facecolor=np.array([255, 192, 3])/255, alpha=.7)
ax.legend(loc=9, ncol=8, fontsize=13, bbox_to_anchor=(.5, 1.07), edgecolor='#FFFFFF')
ax.set_xlim(pd.datetime(2020, 1, 1), pd.datetime(2020, 1, 31))
ax.set_ylim(0, 150)
plt.show()
```
