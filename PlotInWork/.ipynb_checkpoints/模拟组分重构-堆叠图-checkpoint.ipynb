```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 设置中文支持字体
plt.rcParams['font.family'] = 'PingFang HK'
```

## 创建数据集

```python
# # 创建时间序列
# time_ser = pd.date_range('2020-4-1', '2020-4-30', freq='H')
# df = pd.DataFrame(np.random.randint(10, 30, (len(time_ser), 5)), index=time_ser, columns=list('ABCDE'))
# df['total'] = (df.sum(axis=1) + np.random.randn(len(df))).to_numpy().astype(int)
# df

file = r'../data/水溶性离子.xls'
df = pd.read_excel(file, parse_dates={'datetime': [0, 1]}, index_col=0)
df['total'] = df.sum(axis=1) + np.random.randint(1, 5, len(df))
df
```

## 绘制堆叠图

```python
def resetX(time_ser, period, fmt):
    sep = len(time_ser) // period
    return [time.strftime(fmt) for time in time_ser[::sep]]
```

```python
fig = plt.figure(figsize=(10, 6), dpi=120)
# totoal不作为标签
labels = list(df.columns)[:-1]
colors = ['#FF7440', '#4041FF', '#FFBC40', '#A0FF40', '#FF01FF', '#00BFFF', '#A393FF']
ax = fig.add_subplot(111)
ax.stackplot(range(len(df)), *(df[label] for label in labels), labels=labels, colors=colors)
# 绘制totoal
ax.plot(df['total'].to_numpy(), c='#FF2735', label='total')
# 设置legend
ax.legend(loc=9, frameon=False, ncol=len(labels)+1, fontsize=8)
# 设置x轴范围
ax.set_xlim(0, len(df))
# 自定义x标签
ax.set_xticklabels(resetX(df.index, 6, '%Y-%m-%d %H'))
# 隐藏上-右的嵴
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

## 百分比堆叠图

获取百分比数据

```python
data = df.iloc[:, 0:-1]
data = data.div(data.sum(axis=1), axis=0)
data
```

```python
fig = plt.figure(figsize=(10, 6), dpi=120)
ax = fig.add_subplot(111)
ax.stackplot(range(len(df)), *(data[label] for label in labels), labels=labels, colors=colors)
ax.set_xlim(0, len(df))
ax.set_ylim(0, 1)
ax.legend(loc='upper right', bbox_to_anchor=(1.155, 1))
ax.set_xticklabels(resetX(data.index, 6, '%Y-%m-%d %H'))
plt.show()
```
