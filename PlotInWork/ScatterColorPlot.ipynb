记录下绘制带有色标标识的散点图：

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 设置颜色标识，并选择配色

```python
data = np.random.standard_normal((600, 2))
c = np.random.randint(0, 10, len(data))
plt.figure(figsize=(8, 5), dpi=120)
plt.scatter(data[:, 0], data[:, 1], c=c, marker='o', cmap='jet')
plt.colorbar()
plt.title('Scatter Plot')
plt.show()
```

## 加入散点尺寸的数据

```python
data = np.random.standard_normal((600, 2))
c = np.random.randint(0, 10, len(data))
plt.figure(figsize=(8, 5), dpi=120)
# 修改散点的尺寸
size = np.random.randint(5, 20, size=600)
plt.scatter(data[:, 0], data[:, 1], s=size, c=c, marker='o', cmap='jet')
plt.colorbar()
plt.title('Scatter Plot')
plt.show()
```
