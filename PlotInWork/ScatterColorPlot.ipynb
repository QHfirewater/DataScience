记录下绘制带有色标标识的散点图：

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 设置颜色标识，并选择配色

```python
data = np.random.standard_normal((600, 2))
c = np.random.randint(0, 10, len(data))
plt.figure(figsize=(8, 5), dpi=120)
plt.scatter(data[:, 0], data[:, 1], c=c, marker='o', cmap='jet')
plt.colorbar()
plt.title('Scatter Plot')
plt.show()
```

## 加入散点尺寸的数据

```python
data = np.random.standard_normal((600, 2))
c = np.random.randint(0, 10, len(data))
plt.figure(figsize=(8, 5), dpi=120)
# 修改散点的尺寸, 关联一个维度数据
size = np.random.randint(5, 20, size=600)
# 选择系数放大或缩小点的尺寸
plt.scatter(data[:, 0], data[:, 1], s=size*5, c=c, cmap='jet', alpha=.75)
plt.colorbar()
plt.title('Scatter Plot')
plt.show()
```

## 极坐标散点伪彩图

```python
np.random.seed(1960)
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r ** 2
colors = theta

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=.75)
plt.show()
```
